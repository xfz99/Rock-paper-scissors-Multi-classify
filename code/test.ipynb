```python
import cv2
import numpy as np
from skimage import transform
import matplotlib.pyplot as plt
import picture as pic
from PIL import Image
```

```python
from tensorflow.keras.models import load_model


model = load_model('rps16.h5')  #选取自己的.h模型名称

print("finish loading")
```

```python
#单个图片识别
image = cv2.imread('newtestscissors04-13.png')
img=image

print(type(img))
print(img.size)
cv2.imshow("Image1", img)

img=transform.resize(img, (150, 150))
print(type(img))
print(img.size)
img=img.reshape(1,150,150,3)

print(type(img))
print(img.size)

predict = model.predict_classes(img)
print ('识别为：')
if predict==0:
    print ("布")
elif(predict==1):
    print("石头")
elif(predict==2):
    print("剪刀")
else:
    print("其他")

cv2.waitKey(0)
```

```python
#全文件夹识别识别
datapath="scissors"
import os
files= os.listdir(datapath) #得到文件夹下的所有文件名称

for file in files:
    
    filepath=datapath+"/"+file
    #print(file+"被识别为：")
    image = cv2.imread(filepath)
    img=image
    img=transform.resize(img, (150, 150))
    #cv2.imshow("Image1", img)
    img=img.reshape(1,150,150,3)

    predict = model.predict_classes(img)
    #print (predict)
    if predict==0:
        print ("布")
    elif(predict==1):
        print("石头")
    elif(predict==2):
        print("剪刀")
    else:
        print("其他")
    cv2.waitKey(0)
```
